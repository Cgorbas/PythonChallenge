```python
#-The average of the changes in Profit/Losses over the entire period
# use the total from above, n
#  for average, use changeTotal / numRows
def average(numbers):
    length = len(numbers)
    total = 0.0
    for number in numbers:
        total += number
    return total / length


# Test your function with the following:
print(average([1, 5, 9]))
print(average(range(11)))
```
